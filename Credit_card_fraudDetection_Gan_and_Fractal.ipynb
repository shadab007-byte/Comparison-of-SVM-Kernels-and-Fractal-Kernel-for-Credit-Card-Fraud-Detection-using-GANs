## Applying Std scaling to the features before using Generative Adversarial Networks (GANs) is a crucial preprocessing step in this credit card fraud detection study. Here's why this approach is important:

#### 1. Normalization: std scaling normalizes all features to a fixed range (typically 0 to 1), ensuring that no feature dominates due to its larger scale.

#### 2. Improved GAN Performance: GANs, particularly when generating tabular data, benefit from normalized inputs. This scaling helps the generator and discriminator networks learn more effectively.

#### 3. Consistent Feature Representation: By scaling before GAN application, we ensure that the synthetic data generated maintains the same scale and relationships between features as the original data.

#### 4. Preventing Information Leakage: Applying the scaler before data augmentation prevents any potential information leakage between the training and test sets.

#### 5. Stability in Training: Normalized features contribute to more stable gradients during the training of both the GAN and subsequent SVM models.

## The credit card fraud detection dataset underwent Principal Component Analysis (PCA) transformation for several important reasons:

#### Privacy Protection: PCA was applied to maintain client confidentiality by obscuring the original features.

#### Dimensionality Reduction: PCA reduced the dataset to 28 principal components (V1 to V28), potentially simplifying the model without significant loss of information.

#### Multicollinearity Reduction: PCA helps address multicollinearity issues by creating uncorrelated principal components.

#### Effects on further analysis:

#### Interpretability Challenge: The transformed features (V1 to V28) are no longer directly interpretable as specific transaction attributes.

#### Model Performance: PCA may improve model performance by focusing on the most important variance in the data.

#### Computational Efficiency: Reduced dimensionality can lead to faster training and prediction times.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def compare_kernel_performance(X, y):
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Define kernels
    kernels = ['linear', 'poly', 'rbf', 'sigmoid']

    # Create table for results
    results = []

    for kernel in kernels:
        # Train model
        model = SVC(kernel=kernel, probability=True)
        model.fit(X_train, y_train)

        # Make predictions
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_prob)

        results.append({
            'Kernel': kernel,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1,
            'ROC-AUC': auc
        })

    # Convert to DataFrame and display
    results_df = pd.DataFrame(results)
    return results_df.set_index('Kernel').round(4)

# Compare performance of different kernels
performance_comparison = compare_kernel_performance(X, y)
print(performance_comparison)

# Plot metrics comparison
performance_comparison.plot(kind='bar', figsize=(14, 7))
plt.title('Performance Metrics Comparison Across SVM Kernels')
plt.ylabel('Score')
plt.ylim(0, 1.0)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin

class FractalRBF(BaseEstimator, TransformerMixin):
    """
    Fractal-inspired Radial Basis Function transformer.
    Introduces fractal roughness via recursive scaling in the RBF kernel.
    """

    def __init__(self, alpha=0.5, n_iter=3, q=0.75):
        """
        Parameters:
        alpha : float
            Fractal scaling parameter (controls recursive roughness).
        n_iter : int
            Number of recursive iterations to approximate fractal behavior.
        q : float
            Shape parameter controlling the width/locality of the RBF.
        """
        self.alpha = alpha
        self.n_iter = n_iter
        self.q = q

    def _phi(self, r):
        """
        Modified Gaussian RBF function with fractal scaling.
        φ(r) = exp(-(q * r)^2 / (1 + alpha * r^2))
        This formulation slows decay for large r and introduces roughness.
        """
        return np.exp(-(self.q * r)**2 / (1 + self.alpha * r**2))

    def _fractal_basis(self, X, centers):
        """
        Constructs the fractal kernel basis matrix.
        Each basis function is recursively scaled across n_iter iterations.

        Parameters:
        X : ndarray
            Input data of shape (n_samples, n_features).
        centers : ndarray
            Center points selected for basis function evaluation.

        Returns:
        basis : ndarray
            Transformed feature space of shape (n_samples, n_centers).
        """
        basis = np.zeros((X.shape[0], centers.shape[0]))

        # Recursive accumulation over iterations to simulate fractal behavior
        for iter in range(self.n_iter):
            for j, c in enumerate(centers):
                r = np.linalg.norm(X - c, axis=1)  # Euclidean distance
                basis[:, j] += self.alpha**iter * self._phi(r)

        return basis

    def fit(self, X, y=None):
        """
        Randomly selects 100 centers from the input data for the basis functions.
        """
        self.centers_ = X[np.random.choice(X.shape[0], size=100, replace=False)]
        return self

    def transform(self, X):
        """
        Transforms the input data into the fractal RBF feature space.
        """
        return self._fractal_basis(X, self.centers_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define basic fractal RBF function
def fractal_rbf(r, alpha=0.5, q=0.75):
    return np.exp(-(q * r)**2 / (1 + alpha * r**2))

# Create range of r
r = np.linspace(0, 5, 500)

# Plot for different alpha values
plt.figure(figsize=(10, 6))
alphas = [0.0, 0.3, 0.6, 0.9]
for alpha in alphas:
    plt.plot(r, fractal_rbf(r, alpha=alpha), label=f'alpha={alpha}')
plt.title('Effect of alpha on Fractal RBF')
plt.xlabel('r')
plt.ylabel('phi_alpha(r)')
plt.legend()
plt.grid(True)
plt.show()

# Plot for different q values
plt.figure(figsize=(10,6))
qs = [0.5, 1.0, 2.0, 3.0]
for q in qs:
    plt.plot(r, fractal_rbf(r, alpha=0.5, q=q), label=f'q={q}')
plt.title('Effect of q on Fractal RBF')
plt.xlabel('r')
plt.ylabel('phi_alpha(r)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

def fractal_svm(alpha=0.5, q=0.75):
    return Pipeline([
        ('scaler', StandardScaler()),
        ('fractal', FractalRBF(alpha=alpha, q=q)),
        ('svm', SVC(kernel='rbf'))
    ])
```

```python
from sklearn.model_selection import GridSearchCV

def compare_kernels(X, y):
    # Fractal RBF parameters
    param_grid_fractal = {
        'fractal__alpha': [0.3, 0.5, 0.7],
        'fractal__q': [0.5, 0.75, 1.0],
        'svm__C': [0.1, 1, 10]
    }

    # Traditional kernels
    kernels = [
        ('linear', SVC(kernel='linear'), {'C': [0.1, 1, 10]}),
        ('rbf', SVC(kernel='rbf'), {'C': [0.1, 1, 10], 'gamma': ['scale', 'auto']}),
        ('poly', SVC(kernel='poly'), {'degree': [2, 3], 'coef0': [0, 1]}),
    ]

    # Fractal kernel setup
    fractal_model = GridSearchCV(fractal_svm(), param_grid_fractal, cv=5, scoring='f1')

    # Comparison results
    results = []

    # Train and evaluate fractal kernel
    fractal_model.fit(X, y)
    results.append(('fractal', fractal_model.best_score_, fractal_model.best_params_))

    # Train and evaluate traditional kernels
    for name, model, params in kernels:
        gs = GridSearchCV(model, params, cv=5, scoring='f1')
        gs.fit(X, y)
        results.append((name, gs.best_score_, gs.best_params_))

    return results
```

```python
import numpy as np
from sklearn.base import BaseEstimator

class FractalRBF(BaseEstimator):
    def __init__(self, alpha=0.5, n_iter=3, q=0.75, c=1.0):
        self.alpha = alpha  # Fractal scaling parameter (-1 < alpha < 1)
        self.n_iter = n_iter  # Iteration depth (recursive approximation)
        self.q = q          # Shape parameter for RBF
        self.c = c          # Additional RBF control parameter
        self.centers_ = None  # Placeholder for selected centers

    def _hermite_basis(self, r):
        """Hermite polynomial base function as Eq 2.5"""
        return (2*r)**3 - 3*(2*r)  # Cubic Hermite polynomial used to modulate RBF

    def _phi(self, r):
        """Modified RBF with Hermite base integration"""
        # Combines Gaussian-like decay with Hermite basis shape
        return np.exp(-(self.q * r)**2 / (self.c**2 + r)) * self._hermite_basis(r)

    def _fractal_basis(self, X, centers):
        """IFS-based fractal approximation from paper Eq 2.2–2.3"""
        basis = np.zeros((X.shape[0], centers.shape[0]))  # Initialize kernel matrix
        for k in range(self.n_iter):  # Loop over recursive depths
            for j, c in enumerate(centers):  # Loop over each center
                r = np.linalg.norm(X - c, axis=1)  # Compute distance to center
                basis[:,j] += self.alpha**k * self._phi(r / (k+1))  # Apply scaled basis function
        return basis

    def fit(self, X, y=None):
        # Randomly select 100 data points to act as centers (IFS method)
        self.centers_ = X[np.random.choice(X.shape[0], 100, replace=False)]
        return self

    def transform(self, X):
        # Apply the fractal kernel transformation to input X
        return self._fractal_basis(X, self.centers_)

from scipy.linalg import eigh

class MercerFractalRBF(FractalRBF):
    def transform(self, X):
        K = super().transform(X)  # Get kernel matrix from base class

        # Symmetrize and ensure positive semi-definiteness using Mercer correction
        w, V = eigh(0.5*(K + K.T))  # Compute eigenvalues (w) and eigenvectors (V)
        w = np.maximum(w, 1e-6)     # Replace small or negative eigenvalues to ensure stability
        return V @ np.diag(w) @ V.T  # Reconstruct corrected kernel matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

class AlphaFractal:
    def __init__(self, scaling_factors, base_function, trend_function, domain, n_segments, recursion_depth):
        # Initialization of parameters
        self.scaling_factors = scaling_factors      # List of alpha_i (scaling parameters for each transformation)
        self.base_function = base_function          # Function b(x), e.g., Hermite spline
        self.trend_function = trend_function        # Function f(x), e.g., Gaussian RBF
        self.domain = domain                        # Interval [a, b]
        self.n_segments = n_segments                # Number of intervals/subdomains
        self.recursion_depth = recursion_depth      # Depth of recursion for fractal generation
        self.segment_width = (self.domain[1] - self.domain[0]) / self.n_segments  # h = subinterval width
        self.map_list = []                          # Stores the list of transformation functions
        self._generate_transformations()            # Initialize all transformations

    def _generate_transformations(self):
        # Create IFS-based contraction mappings
        for i in range(self.n_segments):
            alpha = self.scaling_factors[i]

            def L(x, i=i):
                # Horizontal mapping into subinterval i
                return (x - self.domain[0]) / self.n_segments + self.domain[0] + i * self.segment_width

            def contract_map(x, y, alpha=alpha, L=L):
                # Fractal transformation in (x, y)
                x_mapped = L(x)
                y_mapped = alpha * (y - self.base_function(x)) + self.trend_function(x_mapped)
                return x_mapped, y_mapped

            self.map_list.append(contract_map)  # Append to transformation list

    def generate_fractal_points(self, x_start, y_start):
        # Recursively generate fractal points from a starting point
        current = [(x_start, y_start)]
        x_points, y_points = [x_start], [y_start]

        for _ in range(self.recursion_depth):
            next_level = []
            for (x, y) in current:
                for transform in self.map_list:
                    x_new, y_new = transform(x, y)
                    x_points.append(x_new)
                    y_points.append(y_new)
                    next_level.append((x_new, y_new))
            current = next_level  # Update for next depth

        return x_points, y_points

    def falpha(self):
        # Build the fractal interpolation function f_alpha(x)
        x_data, y_data = self.generate_fractal_points(0.5, 0.5)

        # Sort x values for interpolation
        sort_indices = np.argsort(x_data)
        x_sorted = np.array(x_data)[sort_indices]
        y_sorted = np.array(y_data)[sort_indices]

        # Keep unique x values to avoid duplicates in interpolation
        unique_idx = np.unique(x_sorted, return_index=True)[1]
        x_unique = x_sorted[unique_idx]
        y_unique = y_sorted[unique_idx]

        # Interpolate fractal function
        f_alpha_func = interp1d(x_unique, y_unique, kind='linear', fill_value="extrapolate")

        return f_alpha_func
```

```python
# Define parameters for the RBF kernel with fractal approximation
domain = [0, np.pi]      # Interval [a, b] over which fractal function is defined
n_segments = 5           # Number of subintervals in IFS
recursion_depth = 9      # Number of recursive iterations for point generation

# Shape parameters for the Gaussian RBF function
c = 1.0                  # Center of Gaussian
s = 0.4                  # Standard deviation

# Define the original Gaussian RBF function
def trend_function(x):
    return np.exp(-((x - c) ** 2) / (2 * s ** 2))

# Derivative of Gaussian RBF for Hermite interpolation
def derivative_of_trend(x):
    return -((x - c) / (s**2)) * trend_function(x)

# Define interpolation nodes for Hermite spline
x_nodes = np.linspace(0, 2 * np.pi, 10)
y_nodes = trend_function(x_nodes)
dy_nodes = derivative_of_trend(x_nodes)

from scipy.interpolate import CubicHermiteSpline
base_function = CubicHermiteSpline(x_nodes, y_nodes, dy_nodes)  # b(x): smooth spline backbone

# Define α values for each IFS transformation (must satisfy |αᵢ| < 1)
scaling_factors = [0.5, -0.5, 0.5, -0.5, 0.5]

# Create the AlphaFractal object and compute the fractal interpolation function f_alpha
alpha_fractal = AlphaFractal(
    scaling_factors=scaling_factors,
    base_function=base_function,
    trend_function=trend_function,
    domain=domain,
    n_segments=n_segments,
    recursion_depth=recursion_depth
)
f_alpha = alpha_fractal.falpha()  # Interpolated fractal function over distance domain

# Function to compute custom fractal kernel matrix using f_alpha(||x - y||)
def fractal_rbf_kernel_matrix(X1, X2, s=1):
    """
    Computes the fractal RBF kernel matrix between two datasets using f_alpha.
    This replaces the standard RBF with a fractal-interpolated distance function.
    """
    gamma = 1 / (2 * (s ** 2))

    # Precompute norms and pairwise squared distances
    norms_1 = np.sum(X1 ** 2, axis=1)[:, np.newaxis]
    norms_2 = np.sum(X2 ** 2, axis=1)[np.newaxis, :]
    distances = norms_1 + norms_2 - 2 * np.dot(X1, X2.T)

    # Apply fractal kernel transformation element-wise
    K = np.zeros((X1.shape[0], X2.shape[0]))  # Initialize kernel matrix
    for i in range(X1.shape[0]):
        for j in range(X2.shape[0]):
            # Compute distance and clip to interpolation domain [0, π]
            scaled_dist = min(np.pi, np.sqrt(np.abs(distances[i, j])))
            # Apply fractal interpolation to scaled distance
            K[i, j] = f_alpha(scaled_dist)

    return K  # Return final fractal kernel matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import CubicHermiteSpline

# Define the AlphaFractal class (same as before)
class AlphaFractal:
    def __init__(self, scale_vector, b_func, f_func, interval, num_intervals, max_depth):
        self.scale_vector = scale_vector
        self.b_func = b_func
        self.f_func = f_func
        self.I = interval
        self.N = num_intervals
        self.max_depth = max_depth
        self.h = (self.I[1] - self.I[0]) / self.N
        self.transformations = []
        self.create_transformations()

    def create_transformations(self):
        for i in range(self.N):
            alpha = self.scale_vector[i]

            def L_i(x, i=i):
                return (x - self.I[0]) / self.N + self.I[0] + i * self.h

            def w(x, y, alpha=alpha, L_i=L_i):
                x_new = L_i(x)
                y_new = alpha * (y - self.b_func(x)) + self.f_func(x_new)
                return x_new, y_new

            self.transformations.append(w)

    def generate_points(self, x0, y0):
        current_points = [(x0, y0)]
        points_x, points_y = [x0], [y0]

        for depth in range(self.max_depth):
            new_points = []
            for (x, y) in current_points:
                for transform in self.transformations:
                    new_x, new_y = transform(x, y)
                    points_x.append(new_x)
                    points_y.append(new_y)
                    new_points.append((new_x, new_y))
            current_points = new_points

        return points_x, points_y

    def falpha(self):
        x_vals, y_vals = self.generate_points(0.5, 0.5)

        sorted_indices = np.argsort(x_vals)
        x_sorted = np.array(x_vals)[sorted_indices]
        y_sorted = np.array(y_vals)[sorted_indices]

        unique_indices = np.unique(x_sorted, return_index=True)[1]
        x_unique = x_sorted[unique_indices]
        y_unique = y_sorted[unique_indices]

        f = interp1d(x_unique, y_unique, kind='linear', fill_value="extrapolate")

        return f

# Define original RBF function and its derivative
def f(x):
    return np.exp(-((x - 1.0) ** 2) / (2 * 0.4 ** 2))

def f_prime(x):
    return -((x - 1.0) / 0.4**2) * f(x)

# Define cubic Hermite spline for interpolation
x_nodes = np.linspace(0, 2*np.pi, 10)
y_nodes = f(x_nodes)
dy_nodes = f_prime(x_nodes)
f_hermite = CubicHermiteSpline(x_nodes, y_nodes, dy_nodes)

# Define interval and number of intervals
I = [0, np.pi]
N = 5
max_depth = 9

# Define alpha sets to visualize
alpha_sets = [
    [0.5, -0.5, 0.5, -0.5, 0.5],  # First alpha set
    [0.3, -0.3, 0.4, -0.4, 0.6],  # Second alpha set
    [0.6, -0.6, 0.7, -0.7, 0.8],  # Third alpha set
]

# Set up the plot
plt.figure(figsize=(12, 8))

# Loop through each alpha set
for alpha_array in alpha_sets:
    # Create the fractal for this alpha set
    alpha_fractal = AlphaFractal(scale_vector=alpha_array, b_func=f_hermite, f_func=f,
                                 interval=I, num_intervals=N, max_depth=max_depth)
    g = alpha_fractal.falpha()

    # Create x-values for plotting
    x_plot = np.linspace(0, np.pi, 1000)

    # Evaluate the fractal function at these x points
    y_plot = g(x_plot)

    # Plot the result for this alpha set
    plt.plot(x_plot, y_plot, label=f'Alpha = {alpha_array}', linewidth=2)

# Add labels, title, and legend
plt.title('Visualization of Alpha-Fractal Function for Different Alpha Sets', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('$f_\\alpha(x)$', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
